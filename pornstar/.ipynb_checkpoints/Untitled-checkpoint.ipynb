```python
face_cascade = cv2.CascadeClassifier('./Source Code/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./Source Code/haarcascades/haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(gray_img, 1.3, 5)
faces
```

```python
#set coordinate 
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
#draw rectangle in the face
face_img = cv2.rectangle(img,(x,y), (x+w, y+h), (100,100,100), 2)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y), (x+w, y+h), (100,100,100), 2)
    roi_gray_img = gray_img[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray_img)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew,ey+ew), (300,300,300),2)

plt.figure()
plt.imshow(face_img, cmap = 'gray')
plt.show()
```

```python
%matplotlib inline
plt.imshow(roi_color, cmap= 'gray')
```

```python
def cropped_images(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
cropped_image = cropped_images('./Dataset\Ai_Uehara\Idol AI Uehara_ poor girl_ Pornstar thong to_yythkg (1).jpg')
cropped_image
```

```python
path_to_data = "./dataset"
path_to_data_crop = "./dataset/crop/"
```

```python
#create path to data
import os
img_dir = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dir.append(entry.path)
```

```python
img_dir
```

```python
#if folder crop exist => remove
import shutil
if os.path.exists(path_to_data_crop):
    shutil.rmtree(path_to_data_crop)
os.mkdir(path_to_data_crop)
```

```python
crop_img_dir = []
pornstar_dict = {}
for img_dr in img_dir:
    count = 1
    pornstar_name = img_dr.split('\\')[-1]
    print(pornstar_name)
    
    for entry in os.scandir(img_dr):
        print(entry.path)
        roi_color = cropped_images(entry.path)
        if roi_color is not None:
            crop_folder = path_to_data_crop + pornstar_name
            if not os.path.exists(crop_folder):
                os.makedirs(crop_folder)
                print('generating foler', crop_folder)                       
            cropped_file_name = pornstar_name + str(count) + ".png"
            cropped_file_path = crop_folder + "/" + cropped_file_name 
            
            cv2.imwrite(cropped_file_path, roi_color)
            pornstar_dict[pornstar_name].append(cropped_file_path)
            count += 1 
```
